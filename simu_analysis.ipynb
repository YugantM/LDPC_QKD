```python
import subprocess
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
main_res = []
for err_prob in np.arange(0.02,0.115,0.01): 
    
    for checks in range(100,750,25):

        err_prob = round(err_prob,3)

        child = subprocess.Popen(['bash','QKD_pipe','{}'.format(err_prob),'{}'.format(checks)],stdout=subprocess.PIPE, stderr=subprocess.PIPE)
        res = child.communicate()[1].decode('utf-8')

        dic = {}
        
        dic['err_prob'] = err_prob
        dic['n_checks'] = checks
        for each in res.split('\n')[4:-1]:

            for _ in each.split(","): 

                if 'Warning' not in _:

                    key,val = _.split(':')
                    dic[key] = float(val)
                    
        print(dic)
        main_res.append(dic)
```

```python
mainframe = pd.DataFrame(main_res)
```

```python
mainframe = pd.read_csv('data.csv')
```

```python
mainframe['1-BLER'] = mainframe['valid blocks']/mainframe['Decoded blocks']
```

```python
def entropy(err_prob):
    
    return -np.log2(err_prob)*err_prob - (1-err_prob)*np.log2(1-err_prob)
```

```python
mainframe['shannon'] = entropy(mainframe['err_prob'])
```

```python
mainframe['real_key'] = 1-mainframe['shannon'] - mainframe['n_checks']/mainframe[' encoded block size']
```

```python
mainframe['eff_key'] = mainframe['1-BLER']*mainframe['real_key']
```

```python
mainframe['eff_key'][mainframe['eff_key']<0] = 0 
```

```python
plt.figure(figsize=(10,6))

plt.xlabel('number of parity')
plt.ylabel('effective key rate')
for x,y in mainframe.groupby('err_prob'):
    
    #print(x,y)
    plt.plot(y['n_checks'],y['eff_key'],'--',label = "{}%".format(round(100*x)))
    
for a,b in mainframe[mainframe['1-BLER']>=1].groupby('err_prob'):
    ana = b.sort_values('n_checks').head(1)
    
    a = ana['eff_key'].tolist()[0]
    b = ana['n_checks'].tolist()[0]
    #print()
    plt.scatter(b,a)
    
plt.legend(title='QBER')
plt.grid()
plt.show()
```
